การสร้าง array แบบ 3 มิติ

```python
mystery_array = np.array([[[0, 1, 2, 3],
                           [4, 5, 6, 7]],
                        
                         [[7, 86, 6, 98],
                          [5, 1, 0, 4]],
                          
                          [[5, 36, 32, 48],
                           [97, 0, 27, 18]]])
```

```python
print(f'We have {mystery_array.ndim} dimensions')
print(f'The shape is {mystery_array.shape}')
```

การดึงข้อมูลจาก array 3 มิติ [ x, y, z] 

x คือ layer, y คือ row, z คือ column

```python
mystery_array[2, 1, 3]
```

```python
mystery_array[2, 1, :]
```

```python
mystery_array[:, :, 0]
```

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])
 
b1 = np.array([[4, 1, 3],
               [5, 8, 5]])
```

```python
c = np.matmul(a1, b1)
print(f"Matrix c has {c.shape[0]} rows and {c.shape[1]} columns")
c
```
