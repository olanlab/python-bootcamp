## Data Visualization

```python
ratings = df_apps_clean.Content_Rating.value_counts()
ratings
```

วิธีการสร้างกราฟ pie ด้วย plotly

สร้างกราฟวงกลมด้วย pie
- labels คือ ป้ายกำกับ
- values คือ ค่า
- title คือ 

```python
print(ratings.index.to_numpy())
print(ratings.values)

fig = px.pie(labels=ratings.index.to_numpy(), values=ratings.values)
fig.show()
```

update_traces คือ การแก้ไขข้อมูล traces
- textposition คือ ตำแหน่งแสดง text
- textinfo คือ การกำหนด text ให้แสดงค่าอะไรบ้าง
- textfont_size คือ ขนาดของตัวอักษร

```python
fig = px.pie(labels=ratings.index.to_numpy(), values=ratings.values, title="Content Rating", names=ratings.index,)
fig.update_traces(textposition='outside', textinfo='percent+label')
 
fig.show()
```

การสร้างกราฟโดนัทด้วย pie
- hole คือ ขนาดรูของโดนัทมีขนาดตั้งแต่ 0 - 1 

```python
fig = px.pie(labels=ratings.index.to_numpy(), values=ratings.values, title="Content Rating", names=ratings.index, hole=0.6,)
fig.update_traces(textposition='inside', textfont_size=15, textinfo='percent')
 
fig.show()
```

การแปลงข้อมูลจาก ข้อความ เป็น ตัวเลข

```python
df_apps_clean.Installs = df_apps_clean.Installs.astype(str).str.replace(',', "")
df_apps_clean.Installs = pd.to_numeric(df_apps_clean.Installs)
df_apps_clean[['App', 'Installs']].groupby('Installs').count()
```

```python
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace('$', "", regex=False)
df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
 
df_apps_clean.sort_values('Price', ascending=False).head(20)
```

```python
df_apps_clean = df_apps_clean[df_apps_clean['Price'] < 250]
df_apps_clean.sort_values('Price', ascending=False).head(5)
```

mul คือ คำสั่งคูณของ DataFrame

```python
df_apps_clean['Revenue_Estimate'] = df_apps_clean.Installs.mul(df_apps_clean.Price)
df_apps_clean.sort_values('Revenue_Estimate', ascending=False)[:10]
```

การแสดงกราฟแท่งด้วย bar
- x คือ ค่าในแกน x
- y คือ ค่าในแกน y
- orientation คือ การแสดงกราฟ แนวนอน หรือ แนวตั้ง

```python
bar = px.bar(x = top10_category.index, y = top10_category.values)
 
bar.show()
```

```python
category_installs = df_apps_clean.groupby('Category').agg({'Installs': pd.Series.sum})
category_installs.sort_values('Installs', ascending=True, inplace=True)

category_installs
```

```python
h_bar = px.bar(x = category_installs.Installs, y = category_installs.index, orientation='h')
h_bar.update_layout(xaxis_title='Number of Downloads', yaxis_title='Category') 

h_bar.show()
```

```python
cat_number = df_apps_clean.groupby('Category').agg({'App': pd.Series.count})

cat_number
```

merge คือ การนำ DataFrame 2 อันมารวมกัน 
- on คือ การเชื่อมกันด้วย column ไหน
- how คือ การเชื่อมกันอย่างไร ประกอบด้วย left, right, outer, inter

```python
cat_merged_df = pd.merge(cat_number, category_installs, on='Category', how="inner")
print(f'The dimensions of the DataFrame are: {cat_merged_df.shape}')
cat_merged_df.sort_values('Installs', ascending=False)
```

scatter
- x คือ ข้อมูลแกน x
- y คือ ข้อมูลแกน y
- title คือ title ของ กราฟ
- size คือ ขนาดของจุด
- hover_name คือ ชื่อเมื่อเรา Mouse ไปวางที่จุด
- color คือ ข้อมูล column ที่ต้องการแยกสี
update_layout
- xaxis_title คือ title ของแกน x
- yaxis_title คือ title ของแกน y
- yaxis คือ การตั้ง Scale ของแกน y

```python
scatter = px.scatter(cat_merged_df, # data
                    x='App', # column name
                    y='Installs',
                    title='Category Concentration',
                    size='App',
                    hover_name=cat_merged_df.index,
                    color='Installs')
 
scatter.update_layout(xaxis_title="Number of Apps (Lower=More Concentrated)",
                      yaxis_title="Installs",
                      yaxis=dict(type='log'))
 
scatter.show()
```

- as_index คือ 
    - True : the key(s) you use in groupby() will become an index in the new dataframe.
    - False : the key(s) you use in groupby() will not become an index in the new dataframe.

```python
print(df_apps_clean.groupby(["Category", "Type"], as_index=True).agg({'App': pd.Series.count}))
print(df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count}))
```

```python
df_free_vs_paid = df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid.head()
```

- barmode คือ mode การแสดงกราฟแท่ง 
    - stack
    - group

```python
g_bar = px.bar(df_free_vs_paid,
               x='Category',
               y='App',
               title='Free vs Paid Apps by Category',
               color='Type',
               barmode='group')
 
g_bar.update_layout(xaxis_title='Category',
                    yaxis_title='Number of Apps',
                    xaxis={'categoryorder':'total descending'},
                    yaxis=dict(type='log'))
 
g_bar.show()
```

การสร้างกราฟ box 
- y คือ ข้อมูลแกน y
- x คือ ข้อมูลแกน x
- color คือ เลือกข้อมูลที่ต้องการแยกสี
- notched คือ notched shape
- points คือ all, outliers
- title คือ title ของกราฟ

update_layout
- yaxis คือ การกำหนด scale ของแกน y

```python
box = px.box(df_apps_clean,
             y='Installs',
             x='Type',
             color='Type',
             notched=True,
             points='all',
             title='How Many Downloads are Paid Apps Giving Up?')
 
box.update_layout(yaxis=dict(type='log'))
 
box.show()
```

```python
df_paid_apps = df_apps_clean[df_apps_clean['Type'] == 'Paid']
box = px.box(df_paid_apps, 
             x='Category', 
             y='Revenue_Estimate',
             title='How Much Can Paid Apps Earn?')
 
box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid App Ballpark Revenue',
                  xaxis={'categoryorder':'min ascending'},
                  yaxis=dict(type='log'))
 
 
box.show()
```

```python
box = px.box(df_paid_apps,
             x='Category',
             y="Price",
             title='Price per Category')
 
box.update_layout(xaxis_title='Category',
                  yaxis_title='Paid App Price',
                  xaxis={'categoryorder':'max descending'},
                  yaxis=dict(type='log'))
 
box.show()
```
