2. Please write a python or pyspark code to re-arrange the data set to item_lv with the given priority
- P0: our_company < CPT
- P1: our_company = CPT
- P2: our_company > CPT

```python
def get_item_lv(competitive_status):
    if(competitive_status == "our_company < CPT"):
        return 'P0'
    elif(competitive_status == "our_company = CPT"):
        return 'P1'
    elif(competitive_status == "our_company > CPT"):
        return "P2"
```

```python
pricing_project_dataset['item_level'] = pricing_project_dataset['our_company_model_competitiveness_status'].apply(get_item_lv)
```
