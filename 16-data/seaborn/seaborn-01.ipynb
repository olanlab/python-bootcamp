```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load sample dataset
titanic = sns.load_dataset("titanic")
titanic
```

```python

# Create a bar plot
sns.barplot(data=titanic, x="class", y="fare", hue="who", errorbar=None)

# Show the plo
plt.show()
```

```python
# Create a facet grid
g = sns.FacetGrid(data=titanic, col="class", row="survived")
g.map(sns.histplot, "age")

# Show the plot
plt.show()
```

```python
# Load sample dataset
flights = sns.load_dataset("flights")
flights
```

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
flights_pivot
```

```python
# Create a heatmap
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Show the plot
plt.show()
```

```python
# Create a time series plot
sns.lineplot(data=flights, x="year", y="passengers", hue="month", marker="o", markersize=5)

# Show the plot
plt.show()
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")
tips
```

```python
# Create a box plot
sns.boxplot(data=tips, x="day", y="total_bill")

# Show the plot
plt.show()
```

```python
# Create a histogram
sns.histplot(data=tips, x="total_bill", bins=10)

# Show the plot
plt.show()
```

```python

# Create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")

# Show the plot
plt.show()
```

```python
# Create a violin plot with swarm points
sns.violinplot(data=tips, x="day", y="total_bill", inner=None)
sns.swarmplot(data=tips, x="day", y="total_bill", color="white", edgecolor="gray")

# Show the plot
plt.show()
```

```python
# Load sample dataset
iris = sns.load_dataset("iris")
iris
```

```python
# Create a pair plot
sns.pairplot(data=iris, hue="species")

# Show the plot
plt.show()
```

```python

# Create a joint distribution plot
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", hue="species")

# Show the plot
plt.show()
```
